Here’s a complete modern workflow to create and manage Python virtual environments using uv.

`uv` is extremely fast and is becoming a preferred choice for modern Python projects, especially for:

* AI/LLM frameworks
* automation platforms
* SaaS backend systems
* CI/CD pipelines

For your test automation and LLM evaluation frameworks, `uv` is actually a very strong choice.

---

# 1. Install UV

Official installation:

## macOS/Linux

```bash id="v2pt1d"
curl -LsSf https://astral.sh/uv/install.sh | sh
```

---

## Windows (PowerShell)

```powershell id="dcxxt5"
powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"
```

---

## Using pip

```bash id="8wv88f"
pip install uv
```

---

## Verify Installation

```bash id="ayp2ud"
uv --version
```

Official website:

[UV Documentation](https://docs.astral.sh/uv/?utm_source=chatgpt.com)

---

# 2. Create a Project Folder

```bash id="6n6eeh"
mkdir ai-test-framework
cd ai-test-framework
```

---

# 3. Create Virtual Environment

## Basic Command

```bash id="3mjlwm"
uv venv
```

This creates:

```text id="2p0r8x"
.venv/
```

inside your project.

---

# 4. Create Environment with Specific Python Version

```bash id="f2m1d9"
uv venv --python 3.11
```

OR

```bash id="0lwfr7"
uv venv --python 3.12
```

---

# 5. Activate the Virtual Environment

## macOS/Linux

```bash id="jlwmpl"
source .venv/bin/activate
```

---

## Windows CMD

```cmd id="g4jlwm"
.venv\Scripts\activate
```

---

## Windows PowerShell

```powershell id="sgjlwm"
.venv\Scripts\Activate.ps1
```

---

# 6. Install Packages

## Using UV

```bash id="v5dg1o"
uv pip install pytest
```

Example:

```bash id="bnhpr5"
uv pip install playwright pytest pandas
```

For your AI automation frameworks:

```bash id="hrgwd7"
uv pip install deepeval langchain openai pytest playwright
```

Install Playwright browsers:

```bash id="8pkzsp"
playwright install
```

---

# 7. Create requirements.txt

```bash id="y6diy5"
uv pip freeze > requirements.txt
```

---

# 8. Install Dependencies from requirements.txt

```bash id="8r3qel"
uv pip install -r requirements.txt
```

---

# 9. Deactivate Environment

```bash id="o9v98x"
deactivate
```

---

# 10. Remove Virtual Environment

Simply delete `.venv`

## macOS/Linux

```bash id="8e7cxz"
rm -rf .venv
```

## Windows

```cmd id="v7u8ra"
rmdir /s .venv
```

---

# 11. Initialize a Full UV Project (Recommended)

Modern approach:

```bash id="r1ifjq"
uv init
```

This creates:

* `pyproject.toml`
* project structure
* dependency management setup

---

# 12. Add Dependencies (Modern Way)

Instead of requirements.txt:

```bash id="x0w7f4"
uv add pytest
```

Example:

```bash id="rr5n39"
uv add playwright pytest deepeval langchain pandas
```

This automatically updates:

* `pyproject.toml`
* lock file

---

# 13. Sync Dependencies

```bash id="cgt74n"
uv sync
```

This installs all dependencies from lock file.

---

# 14. Recommended Modern Workflow

## Create project

```bash id="c4pf8g"
uv init ai-test-framework
cd ai-test-framework
```

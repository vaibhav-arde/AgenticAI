# Virtual environment Creation Using Conda
Here’s a complete step-by-step guide to create and manage a Python virtual environment using Anaconda / Conda.

---

# 1. Install Conda

You can install either:

* Anaconda → Full package distribution with many ML/data science libraries
* Miniconda → Lightweight and recommended for clean setups

Official downloads:

* [Anaconda Download](https://www.anaconda.com/download?utm_source=chatgpt.com)
* [Miniconda Download](https://docs.conda.io/en/latest/miniconda.html?utm_source=chatgpt.com)

---

# 2. Verify Conda Installation

Open terminal and run:

```bash
conda --version
```

Example output:

```bash
conda 25.1.1
```

---

# 3. Update Conda (Recommended)

```bash
conda update -n base -c defaults conda
```

---

# 4. Create a New Virtual Environment

## Basic Syntax

```bash
conda create --name myenv python=3.11
```

Example:

```bash
conda create --name automation-env python=3.11
```

OR shorter syntax:

```bash
conda create -n automation-env python=3.11
```

---

# 5. Activate the Environment

## Windows

```bash
conda activate automation-env
```

## macOS/Linux

```bash
conda activate automation-env
```

After activation you’ll see:

```bash
(automation-env)
```

before terminal prompt.

---

# 6. Install Packages Inside Environment

Example:

```bash
conda install pandas numpy
```

You can also use pip:

```bash
pip install playwright pytest
```

For your automation stack:

```bash
pip install playwright pytest pytest-playwright
```

Install browsers:

```bash
playwright install
```

---

# 7. Check Installed Packages

```bash
conda list
```

---

# 8. Check Available Environments

```bash
conda env list
```

OR

```bash
conda info --envs
```

Example:

```bash
# conda environments:
#
base                  *  /Users/vaibhav/miniconda3
automation-env           /Users/vaibhav/miniconda3/envs/automation-env
```

---

# 9. Deactivate Environment

```bash
conda deactivate
```

---

# 10. Remove/Delete Environment

```bash
conda remove --name automation-env --all
```

---

# 11. Export Environment (Very Important for Teams/Clients)

This is useful for your automation business so clients/team members can reproduce the same setup.

```bash
conda env export > environment.yml
```

---

# 12. Recreate Environment from YAML

```bash
conda env create -f environment.yml
```

---

# 13. Create Environment with Specific Packages

Example for test automation:

```bash
conda create -n qa-framework python=3.11 pytest requests pandas
```

---

# 14. Create Environment in Specific Folder

```bash
conda create --prefix ./env python=3.11
```

Activate:

```bash
conda activate ./env
```

This is useful when shipping frameworks to SaaS clients.

---

# 15. Recommended Setup for Modern Automation Projects

For your Playwright + AI evaluation frameworks:

```bash
conda create -n ai-test-framework python=3.11
conda activate ai-test-framework

pip install playwright pytest deepeval langchain openai pandas

playwright install
```

---

# 16. Useful Conda Commands Cheat Sheet

| Action           | Command                               |
| ---------------- | ------------------------------------- |
| Create env       | `conda create -n myenv python=3.11`   |
| Activate env     | `conda activate myenv`                |
| Deactivate env   | `conda deactivate`                    |
| List envs        | `conda env list`                      |
| Delete env       | `conda remove -n myenv --all`         |
| Install package  | `conda install numpy`                 |
| Install with pip | `pip install pytest`                  |
| Export env       | `conda env export > environment.yml`  |
| Create from yml  | `conda env create -f environment.yml` |
